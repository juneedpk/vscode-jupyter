```python
# importing libraries for gussain mixture model

import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import pandas as pd
import seaborn as sns
```

```python
# creating X and y 

X, y = make_blobs(n_samples=1000, centers=3, random_state=42)

# creating a model

gmm = GaussianMixture(n_components=3)
gmm.fit(X)

# predicting the clustering lables

lables = gmm.predict(X)

# printing first ten labels

print(lables[:10])



```

```python
gmm.get_params()
```

```python
# Check the shape of X
print(X.shape)

# If X has less than 3 columns, reshape it
if X.shape[1] < 3:
    X = np.hstack((X, np.zeros((X.shape[0], 3 - X.shape[1]))))

# Now you can plot the model in plotly
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2], color=lables)
fig.show()
```

```python
df = sns.load_dataset('iris')

# Create a Gaussian Mixture Model with 3 components

iris = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

gmm = GaussianMixture(n_components=3)
gmm.fit(iris)
lables = gmm.predict(X)
print(lables[:10])

# Plot the clusters
fig = px.scatter_3d(x=iris['sepal_length'], y=iris['sepal_width'], z=iris['petal_length'], color=lables)
fig.show()
```

```python
log_likelihood = gmm.score(iris)
print(log_likelihood)
aic = gmm.aic(iris)
print(aic)
bic = gmm.bic(iris)
print(bic)
```
