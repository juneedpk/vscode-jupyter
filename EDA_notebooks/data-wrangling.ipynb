# Data Formatting / Data standardization

- data ko ek standard per lana
- Ensure data is easily understanable
- easy to gather
- easy to workwith
- units will be same for all (units are really important)
- one standard unit for column


```python
#always name in afterwards

kashti1.rename(columns={'age in days': 'ages'},inplace=True)
kashti1.head()

kashti1.rename(columns={'sex': "gender"},inplace=True)
kashti1.head()
```

```python
kashti1=kashti1.drop(kashti1.columns[-1], axis=1)
```

```python
kashti1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

kashti1['gender']= LabelEncoder().fit_transform(kashti1['gender'])
kashti1['class']= LabelEncoder().fit_transform(kashti1['class'])
kashti1['alive']= LabelEncoder().fit_transform(kashti1['alive'])
```

```python
kashti1['age']=np.log(kashti1['age'])
kashti1['fare']=np.log(kashti1['fare'])
```

```python
kashti1.dtypes
```

```python
X = kashti1[['gender','age','class','alive','parch']]
y = kashti1['survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report


model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1).fit(X, y)

y_predict =model.predict(X)

print((classification_report(y, y_predict)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report


model = LogisticRegression().fit(X, y)

y_predict =model.predict(X)

print((classification_report(y, y_predict)))
```

2. Min-Max Method (0 to 1)
3. Z-score (standard score) (-3 to +3) (often called Z score normalization)
4. Log transformation

# Binning

- Grouping of values into smaller number of values (bins)
- convert numeric into categories (young,old,childern) or 1-16, 17-30 etc
- to have better understanding of groups 
     - low vs mid vs high price

```python
bins = np.linspace(min(kashti['age']),max(kashti['age']),15000)

kashti['age']= pd.cut(kashti['age'], bins,include_lowest=True)
kashti['age']
```
