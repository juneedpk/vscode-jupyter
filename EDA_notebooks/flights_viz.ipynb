```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
faa = sns.load_dataset('healthexp')
faa.head()
```

```python
sns.set(style='darkgrid')


sns.pairplot(faa,vars=['Spending_USD','Life_Expectancy'],hue='Country')
plt.show()
```

```python
sns.barplot(x='Country',y='Life_Expectancy',data=faa)
```

```python
sns.barplot(x='Country',y='Spending_USD',data=faa)
```

```python
sns.countplot(faa['Country'])
```

```python


sns.catplot(x='Country',y='Spending_USD',data=faa,col='Year',kind='bar')
plt.figure(figsize=(15,8))
```

```python
sns.histplot(x='Spending_USD',data=faa,hue='Country',kde=True)
```

```python
#faa.groupby(faa['Spending_USD'] > 500).count()

faa.query('Spending_USD < 500').groupby('Country')['Spending_USD'].mean().plot(kind='barh')
```
