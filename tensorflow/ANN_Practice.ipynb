```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
# Importing data set
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1= LabelEncoder()
X[:, 1]= labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2= LabelEncoder()
X[:, 2]= labelencoder_X_2.fit_transform(X[:, 2])


```

```python
# Encoding categorical data now into single cooding to avoid ordinal number perception
onehotencoder= OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()

```

```python
X
```

```python
# Reducing the dummy variables  (Reduce the dummy trap)
X = X[:, 1:]
```

```python
X
```

```python
# Splitting the dataset into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Features Scaling to standardize the data
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler() 
X_train= sc_X.fit_transform(X_train)
X_test= sc_X.transform(X_test)
```

```python
X_test
```

```python
# Start Making ANN
# Import Keras Libraries and Packages
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initiaizing ANN
classifier = Sequential()
```

```python
# Adding the first layer and then hidden layer NEW CHANGES IN KERAS MODEL      here activation is activation fucntion rectifier, unit or output dim is no of nodes in first hidden layer, kernel_initializer means what should be the pattern to assing initial weights quite similar to init in old keras

classifier.add(Dense(activation="relu", input_dim=11, units=6, kernel_initializer="uniform"))
```

```python
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
```

```python
# Adding the Output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN into training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch=100)
```

```python
# Predicting the test results
y_pred= classifier.predict(X_test)
```

```python
y_pred
```

```python
# Decide a threshold which customers will leave== true and which customers wont== false on the basis of probabiliteis in previous cell)
y_pred= (y_pred > 0.5)
```

```python
# Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
