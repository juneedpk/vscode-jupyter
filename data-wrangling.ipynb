```python
ks4= kashti1[['age in days','fare']]
ks4.head()
```

- The above data is really in wide range and we need to normalize and hard to compare
- Normalization change the values to the range of 0 to 1 (now both variables has similar influence on our models)

# Method of Normalization (to make data Human or machine readable size)

1. simple feature scaling 
    . x(new)=x(old)/x(max)
2. Min-Max Method (0 to 1)
3. Z-score (standard score) (-3 to +3) (often called Z score normalization)
4. Log transformation

```python
#simple feature scaling 
ks4['fare']= ks4['fare']/ks4['fare'].max()
ks4['age in days']= ks4['age in days']/ks4['age in days'].max()

ks4.head()
```

```python
# Min-Max method
ks4['age in days'] = (ks4['age in days']-ks4['age in days'].min()) / (ks4['age in days'].max()-ks4['age in days'].min())
ks4.head()
```

```python
# Z score (standard score)  (standardization)
ks4['fare']= (ks4['fare']-ks4['fare'].mean())/ks4['fare'].std()
ks4.head()
```

```python
# log transformation

ks4['fare']= np.log(ks4['fare'])
ks4.head()
```

# Binning

- Grouping of values into smaller number of values (bins)
- convert numeric into categories (young,old,childern) or 1-16, 17-30 etc
- to have better understanding of groups 
     - low vs mid vs high price

```python
bins = np.linspace(min(kashti['age']),max(kashti['age']),15000)

kashti['age']= pd.cut(kashti['age'], bins,include_lowest=True)
kashti['age']
```
